```python
import pandas as pd
import locale
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
locale.setlocale( locale.LC_ALL, 'en_US.UTF-8' )
df = pd.read_json("spider_data_4.json")
```

```python
# Removes commas from yearly sale numbers & total sales numbers
df_2['yearly_sale'] = list(map(locale.atoi,df_2['yearly_sale']))
global_2['yearly_sale'] = list(map(locale.atoi,global_2['yearly_sale']))
usa_2['yearly_sale'] = list(map(locale.atoi,usa_2['yearly_sale']))
japan_2['yearly_sale'] = list(map(locale.atoi,japan_2['yearly_sale']))
europe_2['yearly_sale'] = list(map(locale.atoi,europe_2['yearly_sale']))
usa_2['total_sale'] = list(map(locale.atoi,usa_2['total_sale']))
japan_2['total_sale'] = list(map(locale.atoi,japan_2['total_sale']))
europe_2['total_sale'] = list(map(locale.atoi,europe_2['total_sale']))
```

```python
# Determine the yearly sale's mean to replace N/A values (assumption that yearly sale should = total sale)
global_na = global_2[global_2['total_sale'] == 'N/A']
global_na['yearly_sale'].mean()
```
